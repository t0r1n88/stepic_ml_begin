```python
data = pd.DataFrame({'X_1': [1, 1, 1, 0, 0, 0, 0, 1], 'X_2': [0, 0, 0, 1, 0, 0, 0, 1], 'Y': [1, 1, 1, 1, 0, 0, 0, 0]})
```

```python
data.head()
```

```python
# Создадим экземпляр класса DecisionTreeClassifier и заменим критерий по умолчанию gini на entropy
clf = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
clf
```

```python
# Выделим предикторы(независимые переменные)
# на которых мы будем обучать дерево решений и целевой признак(зависимую переменную)
X = data[['X_1','X_2']]
y = data['Y']
```

```python
# Обучим наше дерево
clf.fit(X,y)
```

```python
# Нарисуем получившееся дерево. С помощью новой функциональности skitlearn
tree.plot_tree(clf.fit(X,y));
```

```python
# Добавим красоты
# filled отвечает за раскраску блоков

tree.plot_tree(clf.fit(X,y),feature_names = list(X),
              class_names=['Negative','Positive'],
              filled=True);
```

```python
tree.plot_tree(clf.fit(X,y),feature_names = list(X),
              class_names=['Negative','Positive'],
              filled=False);
```
